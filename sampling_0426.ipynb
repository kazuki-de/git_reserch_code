```python
def up_sample(path,Split,size,up_path):
    raw = x[x["Split"]==Split]
    raw = raw.sample(frac=1).reset_index(drop=True)
    up = raw[0:size]
    up.to_csv(up_path,index=False)
    return print("finish!")
```
