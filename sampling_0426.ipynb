```python
def add_path(path_x):
    x = pd.read_csv(path_x)
    if "path1" in x.columns:
        x.drop("path1",axis=1)
    z = x.assign(path1=0)
    for i in range(len(x)):
        path = x.loc[i]["file_pass"]
        base = os.path.basename(path)
        path1 = os.path.dirname(path)
        base2 = os.path.basename(path1)
        path2 = os.path.dirname(path1)
        base3 = os.path.basename(path2)
        path3 = os.path.dirname(path2)
        base4 = os.path.basename(path3)
        path4 = os.path.dirname(path3)
        path5 = os.path.join("/mnt/Doc/2048pix_tissue_all",base3,base2,base)
        z["path1"][i] = path5
    z.to_csv(path_x)
    
```
