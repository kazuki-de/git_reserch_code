```python
#n列あるリストの左端列のみを新たなlistに格納する関数
def list_left(list):
    list_a=[]
    for i in range(len(list)):
        a=list[i][0]
        list_a.append(a)
    return list_a
    
```

```python
#ディレクトリからファイル一覧をリストで入手したとき隠しファイルがあると隠しファイルを除去
def listdir_nohidden(path):
    list_n=[]
    for f in os.listdir(path):
        if not f.startswith('.'):
            list_n.append(f)
    return list_n
            
```

```python
#not blunc image sampling
def select_tdel(list_de):
    list_a=[]
    for i in list_de:
        pass1=os.path.join('/home/kazuki-de/Documents/tissue_images',i)
        PP_indev =listdir_nohidden(pass1)
        if len(PP_indev)<50:
            a=random.sample(PP_indev, k=25)
        else:
            a=random.sample(PP_indev, k=50)
        for l in a:
            b=os.path.join(os.path.join(pass1,l))
            list_a.append(b)
    list_d=[]
    for l in list_a:
        img_src = Image.open(l)
        img_temp=img_src
        sd_tmp = np.std(np.asarray(img_src))
        if sd_tmp < 16 :
            continue # skip (almost) blank slice
        list_d.append(l)
    return list_d
```

```python
#file copy to new derectory
def copy_file(filelist,sample_name,type1):
    for ii in filelist:
        for i in ii:
            x=os.path.basename(os.path.dirname(os.path.dirname(i)))
            x1=x[0:23]
            y=os.path.basename(i)
            X=x1+" "+y
            new_pass=shutil.copy(i,os.path.join("/home/kazuki-de/Documents/tissue sample",sample_name,type1,X))
    return new_pass
```

```python
#croped image from 2048*2048 img
def cut_tissue(derectory,size,subtype):
    qqq_src=[]
    name_list=os.listdir(derectory)
    subtype_list=[]
    ID=[]
    for l in name_list:
        nx=ny=size #cuted picture size
        nl=3 #RGB
        nb=10 # number of picture
        a_picture_derectory=os.path.join(derectory,l) #conect file pass
        blunc=np.emptylunc=np.empty((nb, nx, ny, nl), np.float32)     #make blunk arrays
        bb=0
        subtype_list.append(subtype)
        name=l[0:12]
        ID.append(name)
        while(bb<nb):      
            img_src = Image.open(a_picture_derectory) #image open
            w,h=img_src.size
            st_pointw=np.random.choice(w-nx)
            st_pointh=np.random.choice(h-ny)
            cut_img=img_src.crop((st_pointw,st_pointh,st_pointw+nx,st_pointh+ny))
            cut_temp=np.asarray(cut_img)
            img_temp=cut_temp/255.0
            sd_tmp = np.std(np.asarray(cut_temp))
            if sd_tmp < 16 :
                continue # skip (almost) blank slice
            blunc[bb,] = cut_temp
            bb += 1
        
        qqq_src.append(blunc)
    return qqq_src, subtype_list, ID 
```

```python
def remove_smallp(derectory):
    x=os.listdir(derectory)
    list_nodel=[]
    new_del=os.path.join(derectory,"trash")
    os.mkdir(new_del)
    for i in x:
        y=os.path.join(derectory,i)
        img_src=Image.open(y)
        w,h=img_src.size
        if  w==2048 and h==2048:
            list_nodel.append(y)
            shutil.move(y,new_del)
    return list_nodel
```

```python
#TFrecord作成
import glob
import tensorflow as tf
import numpy as np
from PIL import Image
import re
def CreateTFR(path,out_file):
    type_name_list=["PI","PP","TRU"] #
    typeNameList = {"PI":0,"PP":1,"TRU":2}
    files=glob.glob(path)
    labels = 3
    with tf.python_io.TFRecordWriter( out_file ) as writer:
        for i in files:
            # ファイルを開く
            with Image.open(i) as image_object:  # RGB
                ID = re.search(r"[0-9]{2}-[0-9]{4}", i)
                ID_1=ID.group(0)
                #ID_2=re.sub("-", "", ID_1)
                #ID_3=int(ID_2)
                ID_2=bytes(ID_1,"UTF-8")
                image = np.array(image_object)
                h,w,c =image.shape
                image_raw = image.tostring()
                for m in type_name_list:
                    after_words = re.search(m, i)
                    if after_words !=None:
                        label=typeNameList[after_words.group(0)]
                label_1= np.eye(labels)(label)
 
                example = tf.train.Example(features=tf.train.Features(feature={
                        "height": tf.train.Feature(int64_list=tf.train.Int64List(value=[h])),
                        "width": tf.train.Feature(int64_list=tf.train.Int64List(value=[w])),
                        "channel": tf.train.Feature(int64_list=tf.train.Int64List(value=[c])),
                        "label": tf.train.Feature(int64_list=tf.train.Int64List(value=[label])),
                        "ID" : tf.train.Feature(bytes_list=tf.train.BytesList(value=[ID_2])),
                        "image": tf.train.Feature(bytes_list=tf.train.BytesList(value=[image_object.tobytes()]))
                        }))
 
            # レコード書込
            writer.write(example.SerializeToString())                      
#"ID": tf.train.Feature(bytes_list=tf.train.BytesList(value=[ID])),
```
